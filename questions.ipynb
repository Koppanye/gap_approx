## Q3: Are the vertices with $\frac{1}{3}$ not even feasible for the formulation with $=$?

In polymake, the command to use is

`print($p->contains_point(new Vector(1, 2, 0.5)))`

```python
from pyscipopt import Model
model = Model()
model.readProblem("model.lp")
```

```python
from tqdm import tqdm
```

```python
cont = 0
for t, line in enumerate(lines_ineq):
    if "/3" in line:
        x_vec = line.strip().split(" ")
        x_vec = x_vec[1:]
        x_vec_numbers = []
        for x in x_vec:
            if "/" not in x:
                x_vec_numbers.append(float(x))
            else:
                a, b = x.split("/")
                x_vec_numbers.append(float(a) / float(b))
        sol = model.createSol()
        for i, v in enumerate(model.getVars()):
            model.setSolVal(sol, v, x_vec_numbers[i])
        if model.checkSol(sol, printreason=False):
            print(x_vec_numbers)
            break
    if t % 50 == 0:
        print("Done ", t + 1, flush=True)
```
